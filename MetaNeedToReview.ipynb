```python
# https://leetcode.com/problems/exclusive-time-of-functions/description/
# log the process time in result list. Subract the current process time from the previous process time

class Solution:
    def exclusiveTime(self, n: int, logs: List[str]) -> List[int]:

        result = [0]*n
        START = "start"
        END = "end"
        stack = []
        
        for log in logs:
            process, status, ts = log.split(":")
            process, ts = int(process), int(ts)

            if status == START:
                stack.append((process, ts))
            elif status == END:
                popped_process, start_ts = stack.pop()
                elapsed_time = ts - start_ts + 1
                result[popped_process] += elapsed_time
                if stack:
                    result[stack[-1][0]] -= elapsed_time
            else:
                raise ValueError(f"bad status {status}")

        return result

n = 2; logs = ["0:start:0","0:start:2","0:end:5","1:start:6","1:end:6","0:end:7"]
Solution().exclusiveTime(n, logs)
```

```python
# https://leetcode.com/problems/sort-colors
from typing import List

class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        N = len(nums)
        curr, p0, p2 = 0, 0, N-1

        while curr <= p2:
            if nums[curr] == 0:
                nums[p0], nums[curr] = nums[curr], nums[p0]
                curr += 1
                p0 += 1
            elif nums[curr] == 2:
                nums[p2], nums[curr] = nums[curr], nums[p2]
                p2 -= 1
            else:
                curr += 1
        return nums

nums = [2,0,1]
Solution().sortColors(nums)
```

```python
# https://leetcode.com/problems/check-if-an-original-string-exists-given-two-encoded-strings/description/

class Solution:
    def possiblyEquals(self, s1: str, s2: str) -> bool:
        N1, N2 = len(s1), len(s2)

        def possible_nums(s):
            ans = {int(s)}
            for i in range(1, len(s)):
                ans |= {x+y for x in possible_nums(s[:i]) for y in possible_nums(s[i:])}
            return ans
            
        def recur(i, j, diff):
            if i == N1 and j == N2:
                return diff==0

            if i < N1 and s1[i].isdigit():
                end = i
                while end < N1 and s1[end].isdigit():
                    end += 1
                for possible_num in possible_nums(s1[i:end]):
                    if recur(end, j, diff-possible_num):
                        return True
            elif j < N2 and s2[j].isdigit():
                end = j
                while end < N2 and s2[end].isdigit():
                    end += 1
                for possible_num in possible_nums(s2[j:end]):
                    if recur(i, end, diff+possible_num):
                        return True
            elif diff == 0:
                if i<N1 and j<N2 and s1[i] == s2[j]: return recur(i+1, j+1, 0) 
            elif diff > 0:
                if i < N1: return recur(i+1, j, diff-1)
            else:
                if j < N2: return recur(i, j+1, diff+1)
                
            return False

        return recur(0, 0, 0)

s1 = "internationalization"; s2 = "i18n"
Solution().possiblyEquals(s1, s2)
```

```python
# https://leetcode.com/problems/stickers-to-spell-word/

class Solution:
    def minStickers(self, stickers: List[str], target: str) -> int:

        def get_str_signature(strng):
            counter = collections.Counter(strng)
            return [counter.get(chr(ord('a') + i), 0) for i in range(26)]
        
        target = tuple(get_str_signature(target))
        stickers = [get_str_signature(sticker) for sticker in stickers]
        N = len(stickers)        
        cache = {}
        
        def recur(i, target):
            if not any(target):
                # no more stickers needed
                return 0
            if i == N:
                return math.inf

            if (i, target) in cache:
                return cache[(i, target)]

            # do not use
            result = recur(i+1, target)

            # use the sticker
            sticker = stickers[i]
            new_target: List = list(target)
            is_used = False
            
            # print(target)

            for index, count in enumerate(sticker):
                if count and target[index]:
                    is_used = True    
                    new_target[index] -= min(sticker[index], new_target[index])

            if is_used:
                result = min(result, 1 + recur(i, tuple(new_target)))
            
            cache[(i, target)] = result
            return cache[(i, target)]

        min_stickers = recur(0, target)
        
        return min_stickers if min_stickers != math.inf else -1
```

```python
# https://leetcode.com/problems/largest-bst-subtree/?envType=company&envId=facebook&favoriteSlug=facebook-thirty-days

import math

class Solution:
    def largestBSTSubtree(self, root: Optional[TreeNode]) -> int:
      
        def verify(root, min_val=-math.inf, max_val=math.inf):
            if root is None:
                return True
            return (min_val < root.val < max_val) and verify(root.left, min_val, root.val) and verify(root.right, root.val, max_val)

        def count(root):
            if root is None:
                return 0
            return 1 + count(root.left) + count(root.right)

        def find_largest(root):
            if verify(root):
                return count(root)
            
            return max(find_largest(root.left), find_largest(root.right))
            
        return find_largest(root)
```

```python
# https://leetcode.com/problems/multiply-strings

class Solution:
    def multiply(self, num1: str, num2: str) -> str:

        if len(num2) > len(num1):
            return self.multiply(num2, num1)

        def multiply_two_digits(n1: str, n2: str, carry: int=0):
            mul = int(n1) * int(n2) + carry
            return mul//10, mul%10

        def multiply_num_with_digit(n1: str, n2: str):
            if int(n2) ==0:
                return 0
            total = 0
            carry = 0
            for i, digit in enumerate(n1[::-1]):
                carry, mod = multiply_two_digits(digit, n2, carry)
                total = total + mod * 10**i
            if carry:
                total = total + carry * 10**(i+1)
            return total

        total = 0
        for i, digit in enumerate(num2[::-1]):
            if int(digit):
                mul = multiply_num_with_digit(num1, digit) * 10**i
                total += mul
        
        return str(total)

num1="999"; num2="909"
print(Solution().multiply(num1, num2))
print(int(num1) * int(num2))
```

```python
# https://leetcode.com/problems/regular-expression-matching

class Solution(object):
    def isMatch(self, text: str, pattern: str) -> bool:
        if not pattern:
            return not text

        first_match = bool(text) and pattern[0] in {text[0], "."}

        if len(pattern) >= 2 and pattern[1] == "*":
            # a* or .*
            return (
                self.isMatch(text, pattern[2:]) # not match anything
                or
                first_match and self.isMatch(text[1:], pattern) # a* or .*; reuse the pattern
            )
        else:
            return first_match and self.isMatch(text[1:], pattern[1:])
```

```python
# https://leetcode.com/problems/one-edit-distance/description/?envType=company&envId=facebook&favoriteSlug=facebook-thirty-days

class Solution:
    def isOneEditDistance(self, s: str, t: str) -> bool:
        ns, nt = len(s), len(t)
        if s == t or abs(len(s) - len(t)) > 1:
            return False

        if ns > nt:
            return self.isOneEditDistance(t, s)
        
        for i in range(ns):
            if s[i] != t[i]:
                if ns == nt:
                    return s[i+1: ] == t[i+1: ]
                else:
                    return s[i: ] == t[i+1: ]
        
        return True
```

```python
# https://leetcode.com/problems/find-the-celebrity/description
class Solution:
    def findCelebrity(self, n: int) -> int:

        celeb = 0 
        for i in range(1, n):
            if knows(celeb, i):
                celeb = i
        if self.is_celeb(celeb, n):
            return celeb
        return -1

    @staticmethod
    def is_celeb(celeb, n):
        for i in range(n):
            if i == celeb:
                continue
            if knows(celeb, i) or not knows(i, celeb):
                return False
        return True
```

```python
# https://leetcode.com/problems/letter-case-permutation

class Solution:
    def letterCasePermutation(self, s: str) -> List[str]:
        result = []
        N = len(s)
        
        def recur(i, path):
            if i == N:
                result.append("".join(path))
                return

            if s[i].isdigit():
                # keep
                recur(i+1, path + [s[i]])
            else:
                # keep
                recur(i+1, path + [s[i].lower()])
                recur(i+1, path + [s[i].upper()])

        s = list(s)
        recur(0, [])
        
        return result
```

```python
# https://leetcode.com/problems/is-graph-bipartite/description

class Solution:
    def isBipartite(self, graph: List[List[int]]) -> bool:

        N = len(graph)
        color = [0 for _ in range(N)]
        
        def dfs(node, expect_col):
            if color[node] != 0:
                return color[node] == expect_col
            
            color[node] = expect_col
            
            for neigh in graph[node]:
                if dfs(neigh, -expect_col) is False:
                    return False
            return True

        
        for i in range(len(graph)):
            if color[i] == 0:
                if dfs(i, 1) is False:
                    return False
            
        return True
```

```python
# https://leetcode.com/problems/missing-element-in-sorted-array

class Solution:
    def missingElement(self, nums: List[int], k: int) -> int:
        N = len(nums)
        l, r = 0, N-1
        
        while l < r:
            mid = r-(r-l)//2
            if nums[mid] - nums[0] - mid < k:
                l = mid
            else:
                r = mid - 1
        
        return nums[0] + k + l
```

```python
# https://leetcode.com/problems/copy-list-with-random-pointer

"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""

class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':

        if not head:
            return head
        
        old_to_new = {}
        
        def clone(node):
            if node is None:
                return None
            if node not in old_to_new:
                old_to_new[node] = Node(node.val)
            return old_to_new[node]

        old = head
        new = clone(old)

        while old:
            new.random = clone(old.random)
            new.next = clone(old.next)
        
            old = old.next
            new = new.next
        
        return old_to_new[head]
```

```python
# https://leetcode.com/problems/next-permutation/description/?envType=company&envId=facebook&favoriteSlug=facebook-six-months
class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        N = len(nums)
        if N == 1:
            return nums
        if N == 2:
            nums[0], nums[1] = nums[1], nums[0]
            return nums
            
        found = False
        for index in range(N-1, 0, -1):
            if nums[index-1] < nums[index]:
                found = True
                break
        
        if not found:
            nums.sort()
            return nums

        for j in range(N-1, -1, -1):
            if nums[j] > nums[index]:
                break

        nums[j], nums[index] = nums[index], nums[j]
        nums[index+1:] = sorted(nums[index+1:])
```

```python
# https://leetcode.com/problems/convert-binary-search-tree-to-sorted-doubly-linked-list

class Solution:
    def treeToDoublyList(self, root: 'Optional[Node]') -> 'Optional[Node]':

        if not root:
            return root
            
        head = Node(-1)
        prev = head

        def inorder(root):
            nonlocal prev
            if not root:
                return

            inorder(root.left)
            
            prev.right = root
            root.left = prev
            prev = root
            
            inorder(root.right)

        inorder(root)
        head = head.right
        head.left = prev
        prev.right = head

        return head
```

```python
# https://leetcode.com/problems/search-in-rotated-sorted-array

class Solution:
    def binary_search(self, nums, target):
        N = len(nums)
        left, right = 0, N - 1
        
        # index of smallest
        while left <= right:
            mid = (left+right)//2
            if nums[mid] == target:
                return mid
            if target > nums[mid]:
                left = mid + 1
            else:
                right = mid - 1
        return -1

    def find_smallest_index(self, nums):
        N = len(nums)
        left, right = 0, N - 1

        # index of smallest
        while left <= right:
            mid = (left+right)//2
            if nums[mid] > nums[-1]:
                left = mid + 1
            else:
                right = mid - 1
        return left

    def search(self, nums: List[int], target: int) -> int:
        N = len(nums)
        smallest_index = self.find_smallest_index(nums)
        
        index = self.binary_search(nums[:smallest_index], target)
        if index != -1:
            return index

        index = self.binary_search(nums[smallest_index: N+1], target)
        if index != -1:
            return index + smallest_index
        
        return -1
```

```python
# https://leetcode.com/problems/valid-number

class Solution:
    def isNumber(self, s: str) -> bool:

        found_e = False
        found_dot = False
        found_num = False
        N = len(s) 
        for i, char in enumerate(s):
            if not char.isdigit() and (char not in ["+", "-", "e", "E", '.']):
                return False
            if char.isdigit():
                found_num = True
            if char in ["+", "-"]:
                if i != 0 and s[i-1] not in "eE":
                    return False
                if i == N-1:
                    return False
            if char == ".":
                if found_dot or found_e:
                    return False
                found_dot = True
            if char in "eE":
                if found_e or not found_num or i == N-1:
                    return False
                found_e = True

        return found_num
```

```python
# https://leetcode.com/problems/contiguous-array

class Solution:
    def findMaxLength(self, nums: List[int]) -> int:
        
        N = len(nums)
        max_len = 0
        count = 0
        seen_at = {0: -1}
        
        for i, num in enumerate(nums):
            count += 1 if num else -1
            if count in seen_at:
                max_len = max(max_len, i - seen_at[count])
            else:
                seen_at[count] = i
                
        return max_len
```

```python
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:

        N = len(s)
        wordSet = set(wordDict)
        cache = {}
        
        def recur(i):
            if i == N:
                return [""]

            if i in cache:
                return cache[i]

            result = []
            for j in range(i+1, N+1):
                if s[i: j] in wordSet:
                    for next_word in recur(j):
                        result.append(s[i: j] + (" " if next_word else "") + next_word)

            cache[i] = result
            return result

        result = recur(0)
        return result

s = "pineapplepenapple"; wordDict = ["apple","pen","applepen","pine","pineapple"]
Solution().wordBreak(s, wordDict)
```

```python
# https://leetcode.com/problems/insert-delete-getrandom-o1/?envType=company&envId=facebook&favoriteSlug=facebook-six-months

class RandomizedSet:

    def __init__(self):
        self._map = {} #val:indx
        self._arr = [] #val

    def insert(self, val: int) -> bool:
        if val in self._map:
            return False
        else:
            self._map[val] = len(self._arr)
            self._arr.append(val)
            return True

    def remove(self, val: int) -> bool:
        if val not in self._map:
            return False
        
        last_val, last_indx = self._arr[-1], self._map[self._arr[-1]]
        val_indx = self._map[val]
        
        self._arr[val_indx] = last_val
        self._map[last_val] = val_indx
        del self._map[val]
        self._arr.pop()

        return True

    def getRandom(self) -> int:
        return random.choice(self._arr)
```

```python
class Solution:
    def isValidPalindrome(self, s: str, k: int) -> bool:
        N = len(s)
        dp = [[0] * (N) for _ in range(N+1)]

        for start in range(N-2, -1, -1):
            for end in range(start+1, N):
                if s[start] == s[end]:
                    dp[start][end] = dp[start+1][end-1]
                else:
                    dp[start][end] = 1 + min(dp[start][end-1], dp[start+1][end])
        print(dp)
        return dp[0][N-1] <= k
```

```python
# https://leetcode.com/problems/remove-invalid-parentheses

class Solution:
    def removeInvalidParentheses(self, s: str) -> List[str]:

        def find_invalid_counts(s):
            l, r = 0, 0

            for char in s:
                if char == "(":
                    l += 1
                elif char == ")":
                    if l > 0:
                        l -= 1
                    else:
                        r += 1
            return l, r

        def is_valid(s):
            l, r = find_invalid_counts(s)
            return l == r == 0

        def dfs(s, start, l, r):
            if l==0 and r==0 and is_valid(s):
                result.append(s)
                return

            for i in range(start, len(s)):
                if i > 0 and s[i] == s[i-1]:
                    continue
                if r > 0 and s[i] == ")":
                    #delete
                    dfs(s[:i] + s[i+1:], i, l, r-1)
                elif l > 0 and s[i] == "(":
                    #delete
                    dfs(s[:i] + s[i+1:], i, l-1, r)

        l, r = find_invalid_counts(s)
        result = []
        dfs(s, 0, l, r)
        return result
```

```python

class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        # Initialize binary search bounds
        left = 0
        right = len(arr) - k
        
        # Binary search against the criteria described
        while left < right:
            mid = (left + right) // 2
            if x - arr[mid] > arr[mid + k] - x:
                left = mid + 1
            else:
                right = mid

        return arr[left:left + k]
```

```python

# https://leetcode.com/problems/palindrome-partitioning/

from typing import List
class Solution:
    def partition(self, s: str) -> List[List[str]]:

        def is_pallindrome(strng):
            return strng == strng[::-1]

        result = []
        N = len(s)
        
        def create_parts(i, parts):
            if i == N:
                result.append(parts)
                return

            for j in range(i, N):
                if is_pallindrome(s[i:j+1]):
                    create_parts( j+1, parts + [s[i:j+1]])

        create_parts(0, [])
        return result
```

```python
# https://leetcode.com/problems/range-sum-query-2d-immutable

class NumMatrix:

    def __init__(self, matrix: List[List[int]]):
        self.matrix = matrix
        NROWS, NCOLS = len(matrix), len(matrix[0])
        self.matrix_sum = [[0 for _ in range(NCOLS)] for _ in range(NROWS)]

        for i in range(NROWS-1, -1, -1):
            for j in range(NCOLS-1, -1, -1):
                
                total = self.matrix[i][j]
                
                if i+1 < NROWS:
                    total += self.matrix_sum[i+1][j]
                if j+1 < NCOLS:
                    total += self.matrix_sum[i][j+1]
                if i+1 < NROWS and j+1 < NCOLS:
                    total -= self.matrix_sum[i+1][j+1]

                self.matrix_sum[i][j] = total

    def sumRegion(self, row1: int, col1: int, row2: int, col2: int) -> int:
        NROWS, NCOLS = len(self.matrix), len(self.matrix[0])
        total = self.matrix_sum[row1][col1] 
        if col2+1 < NCOLS:
            total -= self.matrix_sum[row1][col2+1]
        if row2+1 < NROWS:
            total -= self.matrix_sum[row2+1][col1]
        if row2+1 < NROWS and col2+1 < NCOLS:
            total += self.matrix_sum[row2+1][col2+1]
            
        return total
```

```python
# https://leetcode.com/problems/binary-tree-maximum-path-sum

class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        max_path_sum = -math.inf
        
        def dfs(root):
            nonlocal max_path_sum
            if root is None:
                return 0

            left_sum = max(dfs(root.left), 0)
            right_sum = max(dfs(root.right), 0)
            max_path_sum = max(max_path_sum, left_sum + right_sum + root.val)
            return max(left_sum, right_sum) + root.val

        dfs(root)
        return max_path_sum
```

```python
# https://leetcode.com/problems/house-robber
class Solution:
    def rob(self, nums: List[int]) -> int:

        # t_minus_two. t_minus_one [0, 1, 2]
        t_minus_two, t_minus_one = 0, 0
        for num in nums:
            temp = max(t_minus_two + num, t_minus_one)
            t_minus_two = t_minus_one
            t_minus_one = temp
        
        return t_minus_one

```

```python
from heapq import *

class MedianFinder:

    def __init__(self):
        self.lo = []
        self.hi = []
        heapify(self.lo) # max_heap  length can be higher
        heapify(self.hi) # min_heap returns smallest numer,

    def addNum(self, num: int) -> None:
        n_lo = len(self.lo)
        n_hi = len(self.hi)
        
        if n_lo == n_hi:
            heappush(self.lo, -heappushpop(self.hi, num))
        else:
            heappush(self.hi, -heappushpop(self.lo, -num))

    def findMedian(self) -> float:
        n_lo = len(self.lo)
        n_hi = len(self.hi)
        
        if n_lo-n_hi==1:
            return -self.lo[0]
        if n_lo == n_hi:
            return (-self.lo[0] + self.hi[0])/2
```

```python
class Solution:
    def hasPath(self, maze: List[List[int]], start: List[int], destination: List[int]) -> bool:

        NROWS, NCOLS = len(maze), len(maze[0])
        visited = set()

        directions = [[0,1], [1, 0], [-1, 0], [0, -1]]
        def dfs(r, c):
            
            if (r, c) in visited:
                return False

            if r == destination[0] and c == destination[1] and maze[r][c] == 0:
                return True
            
            visited.add((r,c))

            for dr, dc in directions:
                nr, nc = r, c
                while 0<=nr+dr<NROWS and 0<=nc+dc<NCOLS and maze[nr+dr][nc+dc] == 0:
                    nr, nc = nr+dr, nc+dc
                if dfs(nr, nc):
                    return True
            return False
        
        return dfs(start[0], start[1])
```
